```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualization
import matplotlib.pyplot as plt # data visualization
```

```python
df = pd.read_csv('../input/fifa-22-complete-player-dataset/players_22.csv')
df.head()
```

```python
def low_number(value):
    color = 'red' if value < 92 else 'black'
    return 'color: %s'%color 
df[['overall','potential']].head().style.applymap(low_number)
```

```python
df.head(2).style.applymap(low_number, subset=['overall', 'potential'])
```

```python
def highlight_max(column):
    is_max = column == column.max()
    return ['background-color: yellow' if x else '' for x in is_max]

df.head(5).style.apply(highlight_max, subset=['age','overall','potential'])
```

```python
df[['overall','potential']].head(5).style.applymap(low_number).apply(highlight_max)
```

```python
# Cria um campo condicional a outro
df['age_class'] = np.where(df['age']>30, 'old', 'young') # import numpy as np
df['age_class']
```
